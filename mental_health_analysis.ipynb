```python
import seaborn as sns
import matplotlib.pyplot as plt

stress_order = ["Low", "Medium", "High", "Very High"]

sns.boxplot(data=df,x="Most_Used_Platform",y="Stress_Level",orient="v",color="yellow")
plt.title("Stress level by most used platform")
plt.tight_layout()
plt.xticks(rotation = 75)
plt.yticks(range(len(stress_order)), stress_order)
plt.show()
```

```python
pd.crosstab(
    df["Most_Used_Platform"],
    df["Stress_Level"],
    normalize="index"
) * 100
```

```python
import pandas as pd
import matplotlib.pyplot as plt

stress_pct = pd.crosstab(
    df['Most_Used_Platform'],
    df['Stress_Level'],
    normalize='index'
) * 100

stress_pct = stress_pct[['Low', 'Medium', 'High', 'Very High']]
stress_pct.plot(
    kind='bar',
    stacked=True,
    figsize=(12, 6),
    colormap='viridis'
)

plt.title('Stress Level Distribution by Most Used Platform (%)')
plt.xlabel('Most Used Platform')
plt.ylabel('Percentage of Users')
plt.xticks(rotation=45)
plt.legend(title='Stress Level', loc='upper left')
plt.tight_layout()

plt.show()
```

```python
sns.regplot(data=df,x= "Mental_Health_Score",y="Sleep_Hours_Per_Night",color="red", line_kws={"color": "blue", "linewidth": 2,"linestyle":"--","alpha":0.6})
plt.title("Sleeping hours by mental health score")
plt.tight_layout()
plt.show()
```

```python
g_st = pd.crosstab(
    df["Gender"],
    df["Stress_Level"],
    normalize="index"
) * 100

g_st = g_st[["Low", "Medium", "High", "Very High"]]

g_st.plot(
    kind="bar",
    stacked=True,
    colormap="rainbow"
)

plt.title("Gender-wise Stress Level Distribution")
plt.xlabel("Gender")
plt.ylabel("Percentage (%)")
plt.tight_layout()
plt.show()
```

```python
a_st_slp = pd.crosstab(
    index=df["Age"],
    columns=df["Stress_Level"],
    values=df["Sleep_Hours_Per_Night"],
    aggfunc="mean"
)

a_st_slp = a_st_slp[["Low", "Medium", "High", "Very High"]]

plt.figure(figsize=(10,6))
sns.heatmap(a_st_slp, annot=True, cmap="viridis")

plt.title("Age vs Stress Level (Avg Sleep Hours)")
plt.xlabel("Stress Level")
plt.ylabel("Age")
plt.tight_layout()
plt.show()
```
